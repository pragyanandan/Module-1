# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
# A:
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(df['age'], bins=30, alpha=0.5, edgecolor='black', label='Age')
for patch in patches:
    patch.set_width(patch.get_width() * 0.9)
    patch.set_x(patch.get_x() + patch.get_width() * 0.05)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.legend(loc='upper right')
plt.show()
````

#### 4.B BoxPlot of `age`

````python
# A:
plt.figure(figsize=(10, 6))

plt.boxplot(df['age'], vert=False, patch_artist=True, labels=['Height'])

plt.ylabel('Value')
plt.title('Box Plot of Age')

plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
# A:
# Create a new figure
plt.figure(figsize=(10, 6))

plt.scatter(df['age'], df['length_of_service'], alpha=0.5, c='red', edgecolors='w', s=20)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Length of service (year)')
plt.title('Scatter Plot of Age and Length of Service')
````
