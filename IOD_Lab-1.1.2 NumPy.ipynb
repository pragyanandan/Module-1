## Summary statistics with height, weight and population data

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

Find the dimension of each array: height, weight, population
```

```

````python
print("np.ndim(height) : ", height.ndim)
print("np.ndim(weight) : ", weight.ndim)
print("np.ndim(population) : ", population.ndim)
````

**10. Create histograms of height and weight**

````python

plt.figure(figsize=(10, 6))
plt.hist(weight, bins=30, alpha=0.5, label='Weights')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Weights')
plt.legend(loc='upper right')
plt.show()
````

**11. Create separate boxplots of height and weight**

````python
help(plt.boxplot)
````

````python


# Plot box plot
plt.figure(figsize=(10, 6))

plt.boxplot(height, vert=False, patch_artist=True, labels=['Height'])

plt.ylabel('Value')
plt.title('Box Plot of Height')

plt.show()
````

A box plot of the data set can be generated by first calculating five relevant values of this data set: minimum, maximum, median (Q2), first quartile (Q1), and third quartile (Q3).
the first quartile value (Q1 or 25th percentile) 
The third quartile value (Q3 or 75th percentile) 


````python
# Plot box plot
plt.figure(figsize=(10, 6))

plt.boxplot(weight, vert=False, patch_artist=True, labels=['Weight'])

plt.ylabel('Value')
plt.title('Box Plot of Weight')

plt.show()
````

**12. Create a scatterplot of height and weight**

````python
help(plt.scatter)
````

````python




plt.figure(figsize=(10, 6))

plt.scatter(height, weight, alpha=0.5, c='red', edgecolors='w', s=20)

# Add labels and title
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot of Height and Weight')


# Show the plot
plt.show()
````

What is Hexbin plot


````python
plt.figure(figsize=(10, 6))

# Plot hexbin plot
plt.hexbin(height, weight, gridsize=30, cmap='Blues', mincnt=1)
cb = plt.colorbar(label='Count')

# Add labels and title
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.title('Hexbin Plot of Height and Weight')

# Show the plot
plt.show()
````
