````python
# Calculate the coefficient of best fit line or curve
from scipy.stats import linregress
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Coefficient of determination (R-squared)
r_squared = r_value**2

# Print results
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_squared)

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

plt.plot(x, y, 'o', label='original data')
plt.plot(x, (intercept + slope*np.array(x)), 'r', label='fitted line')
plt.legend()
plt.show()
````

````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    n = len(X)
    sum_x = sum(X)
    sum_y = sum(Y)
    sum_x_squared = sum(x ** 2 for x in X)
    sum_xy = sum(X[i] * Y[i] for i in range(n))
    
    #Slope calculation
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
   
   #Intercept calculation
    b = (sum_y - m * sum_x) / n

    return m, b

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X, Y, color='blue', label='data')

# Plot best fit line
plt.plot(X, [m * x + b for x in X], color='red', label='Best Fit Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
````

# 13. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m = np.mean(a)
  se = stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))

results = stats.ttest_1samp(data,170,alternative='greater')
````
